```python
# Mapping of Bengali digits to Western digits
BENGALI_DIGITS = {
    '০': '0', '১': '1', '২': '2', '৩': '3', '৪': '4',
    '৫': '5', '৬': '6', '৭': '7', '৮': '8', '৯': '9'
}

# Mapping of Bengali month names to month numbers (1‑12)
BENGALI_MONTHS = {
    'জানুয়ারি': 1, 'ফেব্রুয়ারি': 2, 'মার্চ': 3, 'এপ্রিল': 4,
    'মে': 5, 'জুন': 6, 'জুলাই': 7, 'আগস্ট': 8,
    'সেপ্টেম্বর': 9, 'অক্টোবর': 10, 'নভেম্বর': 11, 'ডিসেম্বর': 12
}
# Add any alternative spellings if necessary

def parse_bengali_date(date_str):
    """
    Convert a Bengali date string like
    "বৃহস্পতিবার, ০৫ ফেব্রুয়ারি ২০২৬ ২২ মাঘ ১৪৩২"
    to a pandas Timestamp (Gregorian).
    """
    # Split the string by spaces
    parts = date_str.split()

    # Gregorian parts: after the weekday, at indices 1, 2, 3
    day_bn = parts[1].strip()        # e.g., "০৫"
    month_bn = parts[2]               # e.g., "ফেব্রুয়ারি"
    year_bn = parts[3]                 # e.g., "২০২৬"

    # Convert Bengali digits to integers
    day = int(''.join(BENGALI_DIGITS[ch] for ch in day_bn))
    year = int(''.join(BENGALI_DIGITS[ch] for ch in year_bn))
    month = BENGALI_MONTHS[month_bn]

    # Return a Timestamp
    return pd.Timestamp(year, month, day)
```

```python
df_banglaTribune['date'] = df_banglaTribune['date'].apply(parse_bengali_date)
```
